**1) Using inbuilt SVD function in python**

Load dataset

```python
from sklearn.datasets import load_digits
digits=load_digits()
X=digits.data
y=digits.target
print(X.shape)
```

```python
print(y)
```

Lets try it with first image of dataset.

```python
from matplotlib import pyplot as plt
image=X[0]
image=image.reshape((8, 8))
plt.matshow(image,cmap='gray')
```

image=U.E.(VT)

```python
import numpy as np
U,E,VT=np.linalg.svd(image)
U.shape, VT.shape, E.shape
print("U:",U)
print("VT:",VT)
print("E:",E)
```

Reducing dimention to 2

Reduced matrix=UE=AV=A.((V.T).T)

```python
Sigma=np.diag(E)
print("Sigma\n",Sigma)
n_elements = 2
Sigma=Sigma[:,:n_elements]
VT=VT[:n_elements,:]
B=U.dot(Sigma.dot(VT))     # B=U.E.(V.T)=image=X[0]
print("B:\n",B)
T=U.dot(Sigma)
print("U.E\n",T)
T=image.dot(VT.T)
print("A.V:\n",T)
```

```python
plt.matshow(B,cmap = 'gray')
```

**2) Step-by-step**

```python
A=image
A.shape
```

Get eigen values, eigen vectors of A.(A.T)

```python
import scipy.linalg as la
values, vectors = la.eig(A.dot(A.T))
sigma=np.sqrt(values)
sigma[6]=0
print("Eigen values or E:\n",sigma)
print("Eigen Vector or U:\n",vectors)   #vectors is same as U
```

Check for U:  U.(U.T) is supposed to be equal to I(8x8)

```python
print(vectors.dot(vectors.T))
```

Reduced dimentionality

U.E=A.V

```python
Sigma=np.diag(sigma)
n_elements=2
Sigma=Sigma[:,:n_elements]
T=vectors.dot(Sigma)
print("U.E:\n",T)
```

# Q2

(a) Load the dataset in a dataframe.

```python
import pandas as pd
url='/content/drive/MyDrive/ML assignments/Assignment 3/IMDB Dataset.csv'
imdb=pd.read_csv(url)
print(imdb)
```

(b) Drop the sentiment column and consider the first 1000 reviews.

```python
imdb=imdb.drop(columns=['sentiment'])
print(imdb)
```

(c) Make the corpus of first 1000 reviews.

```python
imdb2=imdb.head(1000)
print(imdb2)
```

(d) Convert the corpus into binary BOW vector of size mXn, where m=1000 (number of reviews
documents) and n is the number of unique terms obtained from the 1000 documents. Each ij th entry
of the vector is a binary value which is 1 if the jth term is present in ith review else 0.

```python
corpus=[]
for i in range(len(imdb2['review'])):
  corpus.append(imdb['review'][i])
print(len(corpus))
```

(e) Compute the co-occurrence matrix of order nXn where each ij th entry of matrix is number of
documents in which both i and j th terms co-occur. (Use binary co-occurrence vector to compute
it).

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
vectorizer.fit(corpus)
vector = vectorizer.transform(corpus)
print("Bag of Words:\n",vectorizer.vocabulary_)
A=vector.toarray()
print("Binary Matrix:\n",A)
```

```python
print(A.shape)
```

```python
features=A.T
cov_matrix = np.cov(features)
print("Co-occurance Matrix:\n",cov_matrix)
```

```python
print(cov_matrix.shape)
```

(f) Using Truncated SVD method of python find the reduced matrix of co-occurrence matrix with
number of coomponents as 100.

```python
import sklearn.decomposition as skd
svd=skd.TruncatedSVD(n_components=100)
svd.fit(cov_matrix)
result=svd.transform(cov_matrix)
print(result)
```

(g) The reduced matrix of order nX100 are word embeddings of n words of dimensionality 100.

```python
print(result.shape)
```

**Dimensionality reduction via linear discriminant analysis**

Read dataset

```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
X=pd.DataFrame(iris.data, columns=iris.feature_names)
y=pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
X.shape
```

```python
X.head()
```

```python
iris.target_names
```

Encode categorical class labels

```python
from sklearn.preprocessing import LabelEncoder
class_le=LabelEncoder()
y=class_le.fit_transform(y)
```

```python
y
```

Standardize features

```python
from sklearn.preprocessing import StandardScaler
stdsc=StandardScaler()
X_train_std=stdsc.fit_transform(X)
```

```python
X_train_std
```

Construct within-class covariant scatter matrix S_within

```python
S_within=np.zeros((4,4))
for i in range(3):
  S_within+=np.cov(X_train_std[y==i].T)
```

```python
S_within
```

Construct between-class scatter matrix S_B

```python
N=np.bincount(y) 
vecs=[]
[vecs.append(np.mean(X_train_std[y==i],axis=0)) for i in range(3)] 
mean_overall=np.mean(X_train_std, axis=0) 
S_between_class=np.zeros((4,4))
for i in range(3):
  z=vecs[i]-mean_overall
  S_between_class+=N[i]*((z.reshape(4,1)).dot((z.reshape(1,4))))
```

```python
S_between_class
```

Calculate sorted eigenvalues and eigenvectors of inverse(S_within)dot(S_between_class)

```python
eigen_vals, eigen_vecs=np.linalg.eig(np.linalg.inv(S_within).dot(S_between_class))
eigen_pairs=[(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(len(eigen_vals))]
eigen_pairs=sorted(eigen_pairs,key=lambda k: k[0], reverse=True)
print('Eigenvalues in decreasing order:\n')
for eigen_val in eigen_pairs:
  print(eigen_val[0])
```

Plot main LDA components

```python
import matplotlib.pyplot as plt
tot=sum(eigen_vals.real)
discr=[(i / tot) for i in sorted(eigen_vals.real, reverse=True)]
cum_discr=np.cumsum(discr)
plt.bar(range(1, 5), discr, width=0.2,alpha=0.5, align='center',label='individual "discriminability"')
plt.step(range(1, 5), cum_discr, where='mid',label='cumulative "discriminability"')
plt.ylabel('"discriminability" ratio')
plt.xlabel('Linear Discriminants')
plt.ylim([-0.1, 1.1])
plt.legend(loc='best')
plt.show()
```

Project original features onto the new feature space

```python
W=np.hstack((eigen_pairs[0][1][:, ].reshape(4,1),eigen_pairs[1][1][:, ].reshape(4,1))).real
X_train_lda=X_train_std.dot(W)
```

Plot transformed features in LDA subspace

```python
data=pd.DataFrame(X_train_lda)
data['class']=y
data.columns=["LD1","LD2","class"]
data
```

```python
import seaborn as sns
markers = ['s', 'x','o']
sns.lmplot(x="LD1", y="LD2", data=data, markers=markers,fit_reg=False, hue='class', legend=False)
plt.legend(loc='upper center')
plt.show()
```
