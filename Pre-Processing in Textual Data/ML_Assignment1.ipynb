# **Normalisation**

# **BoW**

Method 1 *Using CountVectorizer*

```
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(ngram_range=(1,1))   #ngrams represents the combination of words to be defined as singla differential entity
BoW=[]                                          #would contain bag of words
feature_matrix=[]
data1=[]
for i in range(len(lower_corpus)):
  input=lower_corpus[i]
  input=[input]
  data1.append(input)                           #data1 now contain data as list of lists.

for i in range(len(data1)):
  vectorizer.fit(data1[i])
  vector=vectorizer.transform(data1[i])
  BoW.append(vectorizer.vocabulary_)
  feature_matrix.append(vector.toarray())
print("Feature Matrix:")
for i in range(len(feature_matrix)):
  for j in range(len(feature_matrix[i])):
    print(feature_matrix[i][j])
  print("\n")
print("Bag of Word:\n",BoW)
```

Without in-built function TfidfVectorizer

```
import math
def idf(corpus,unique_words):
  idf_dict={}
  n=len(corpus)
  for i in unique_words:
    count=0
    for text in corpus:
      if i in text.split():
        count=count+1
      idf_dict[i]=(math.log((1+n)/(count+1)))+1      #extra 1 for avoiding denominator eing zero
  return idf_dict

def fit(whole_data):
  unique_words=set()
  if isinstance(whole_data,(list,)):
    for x in whole_data:
      for y in x.split():
        if len(y)<2:                                 #removing words of size<2
          continue
        unique_words.add(y)
        vocab = {j:i for i,j in enumerate(unique_words)}
        Idf_values_of_all_unique_words=idf(whole_data,unique_words)
  return vocab, Idf_values_of_all_unique_words
Vocabulary, idf_of_vocabulary=fit(data)
print(list(Vocabulary.keys()))
print(list(idf_of_vocabulary.values()))
```

```
from scipy.sparse import csr_matrix
import collections
import numpy as np
from sklearn.preprocessing import normalize
def transform(dataset,vocabulary,idf_values):
  sparse_matrix=csr_matrix((len(dataset),len(vocabulary)),dtype=np.float64)
  for row in range(0,len(dataset)):
    word_count=collections.Counter(dataset[row].split())
    for word in dataset[row].split():
      if word in list(vocabulary.keys()):
        tf_idf_val=(word_count[word]/len(dataset[row].split()))*idf_values[word]
        sparse_matrix[row,vocabulary[word]]=tf_idf_val
  print("Normal Form")
  print(normalize(sparse_matrix, norm='l2', axis=1, copy=True, return_norm=False))
  output=normalize(sparse_matrix, norm='l2', axis=1, copy=True, return_norm=False)
  return output
final_output=transform(data,Vocabulary,idf_of_vocabulary)
print(final_output)
```
